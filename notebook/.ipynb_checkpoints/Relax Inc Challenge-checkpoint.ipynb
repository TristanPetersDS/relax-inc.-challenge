# 1. Data Loading and Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
engagement_df = pd.read_csv('takehome_user_engagement.csv')
users_df = pd.read_csv('takehome_users.csv', encoding='latin-1')

# Rename 'object_id' to 'user_id' in users_df for consistency
users_df.rename(columns={'object_id': 'user_id'}, inplace=True)

# Convert timestamp columns to datetime objects
engagement_df['time_stamp'] = pd.to_datetime(engagement_df['time_stamp'])
users_df['creation_time'] = pd.to_datetime(users_df['creation_time'])
users_df['last_session_creation_time'] = pd.to_datetime(users_df['last_session_creation_time'], unit='s')


print("User Engagement Data:")
display(engagement_df.head())
print("\nUser Data:")
display(users_df.head())
```

# 2. Defining "Adopted User"

```python
def is_adopted_user(user_engagement):
    """
    Checks if a user is an "adopted user" based on their login history.
    An adopted user is defined as a user who has logged in on three separate days
    in a seven-day period.
    """
    # Sort by timestamp to ensure chronological order
    user_engagement = user_engagement.sort_values('time_stamp')
    
    # Drop duplicate days, as we are looking for logins on separate days.
    user_engagement = user_engagement.drop_duplicates(subset='time_stamp', keep='first')
    
    # If the user has less than 3 logins, they can't be an adopted user
    if len(user_engagement) < 3:
        return 0

    # Use a sliding window of size 3 to check for 3 logins within 7 days
    for i in range(len(user_engagement) - 2):
        # Get the timestamps of 3 consecutive logins
        start_date = user_engagement.iloc[i]['time_stamp']
        end_date = user_engagement.iloc[i+2]['time_stamp']
        
        # Check if the time difference is within 7 days
        if (end_date - start_date).days <= 7:
            return 1
            
    return 0

# Group by user and apply the function
adopted_status = engagement_df.groupby('user_id').apply(is_adopted_user, include_groups=False)
adopted_status.name = 'adopted_user'

# Merge the adopted_user status with the users dataframe
users_df = users_df.merge(adopted_status, on='user_id', how='left')
users_df['adopted_user'] = users_df['adopted_user'].fillna(0) # Fill NaN with 0 for users with no engagement data

print("\nUser Data with Adopted User feature:")
display(users_df.head())
print(f"\nNumber of adopted users: {users_df['adopted_user'].sum()}")
```

# 3. Exploratory Data Analysis
Now, let's explore the data to understand the characteristics of adopted vs. non-adopted users.

## Adopted User Distribution

```python
# Plot the distribution of adopted users
plt.figure(figsize=(6, 4))
sns.countplot(x='adopted_user', data=users_df)
plt.title('Distribution of Adopted Users')
plt.xlabel('Adopted User (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

## Creation Source vs. Adoption

```python
# Plot the relationship between creation source and adoption
plt.figure(figsize=(10, 6))
sns.countplot(x='creation_source', hue='adopted_user', data=users_df)
plt.title('Adoption Rate by Creation Source')
plt.xlabel('Creation Source')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Adopted User', labels=['No', 'Yes'])
plt.show()
```

## Mailing List and Marketing Drop vs. Adoption

```python
# Plot the relationship between mailing list opt-in and adoption
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='opted_in_to_mailing_list', hue='adopted_user', data=users_df)
plt.title('Adoption Rate by Mailing List Opt-in')
plt.xlabel('Opted-in to Mailing List')
plt.ylabel('Count')

# Plot the relationship between marketing drip and adoption
plt.subplot(1, 2, 2)
sns.countplot(x='enabled_for_marketing_drip', hue='adopted_user', data=users_df)
plt.title('Adoption Rate by Marketing Drip')
plt.xlabel('Enabled for Marketing Drip')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

# 4. Feature Engineering

```python
# Account Age
users_df['account_age_days'] = (users_df['last_session_creation_time'] - users_df['creation_time']).dt.days

# Fill missing account ages with the median
median_age = users_df['account_age_days'].median()
users_df['account_age_days'] = users_df['account_age_days'].fillna(median_age)

# Encode categorical variables
users_df = pd.get_dummies(users_df, columns=['creation_source'], drop_first=True)

# Drop columns that are not useful for modeling
users_df = users_df.drop(['user_id', 'name', 'email', 'creation_time', 'last_session_creation_time', 'org_id', 'invited_by_user_id'], axis=1)

print("\nFinal preprocessed data for modeling:")
display(users_df.head())
```

# 5. Predictive Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Define features (X) and target (y)
X = users_df.drop('adopted_user', axis=1)
y = users_df['adopted_user']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Feature Importance
feature_importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances, y=feature_importances.index)
plt.title('Feature Importances')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```

# 6. Conclusion and Recommendations
The Random Forest model performs reasonably well in predicting user adoption. The most important predictive factors are:

- **Account Age**: The age of a user's account is the strongest predictor. This suggests that users who have been on the platform longer are more likely to become adopted users.

- **Creation Source**: How a user signs up also plays a significant role.

- **Marketing Drip and Mailing List**: While EDA didn't show a strong visual correlation, the model finds some predictive power in these features.

Based on these findings, here are some recommendations:

- **Focus on Early Engagement**: Since account age is a key factor, focus on engaging new users within their first few weeks to increase their chances of becoming adopted users.

- **Optimize Invitation Channels**: Analyze the GUEST_INVITE and ORG_INVITE channels to understand what makes them effective and apply those learnings to other channels.

- **Personalize Marketing**: Even though their impact is smaller, the marketing features are still relevant. Consider personalizing marketing campaigns to better target users and encourage adoption.

This analysis provides a solid foundation for understanding user adoption. Further investigation could involve a more in-depth analysis of user behavior within organizations and the impact of the users who invite others.
